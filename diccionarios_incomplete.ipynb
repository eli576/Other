```python
string1 = "hello"
string2 = "world"

# Compare the first character of both strings
if string1[0] == string2[0]:
    print(f"The first characters '{string1[0]}' and '{string2[0]}' are the same.")
else:
    print(f"The first characters '{string1[0]}' and '{string2[0]}' are different.")

# Compare the second character of both strings
if string1[1] == string2[1]:
    print(f"The second characters '{string1[1]}' and '{string2[1]}' are the same.")
else:
    print(f"The second characters '{string1[1]}' and '{string2[1]}' are different.")

# You can also loop through and compare characters at each index
print("\nComparing characters at each index:")
min_len = min(len(string1), len(string2))
for i in range(min_len):
    if string1[i] == string2[i]:
        print(f"Index {i}: '{string1[i]}' and '{string2[i]}' are the same.")
    else:
        print(f"Index {i}: '{string1[i]}' and '{string2[i]}' are different.")
```

### Explanation:

1.  **Accessing Characters**: Strings in Python can be treated like arrays of characters, and you can access individual characters using square brackets `[]` with an index. The index starts from `0` for the first character.
    *   `string1[0]` gives you the first character of `string1`.
    *   `string2[1]` gives you the second character of `string2`.

2.  **Comparison**: Once you have the characters, you can compare them directly using the equality operator `==`.

3.  **Looping for Multiple Comparisons**: To compare characters at multiple positions, you can use a `for` loop that iterates over the range of indices common to both strings (using `min(len(string1), len(string2))` to avoid `IndexError` if strings are of different lengths).
